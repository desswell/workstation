```python
def opt_xgb(input_data, gens: list) -> float:

  param = {"max_depth": int(gens[1]), "eta": gens[0], "gamma": gens[2], "alpha": gens[3], "lambda": gens[4], "objective": "binary:logistic", 'verbosity': 0 }
  res = xgb.cv(
    param,
    input_data,
    num_boost_round=10000,
    nfold=5,
    metrics={"logloss"},
    seed=19,
    callbacks=[xgb.callback.EvaluationMonitor(show_stdv=False), xgb.callback.EarlyStopping(5)],
)
  return res['test-logloss-mean'][len(res)-1]
```
