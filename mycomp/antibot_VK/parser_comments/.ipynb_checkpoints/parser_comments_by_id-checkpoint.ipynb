```python
import pandas as pd
```

```python
df = pd.read_csv('010122_220222.csv')
urls = df['url'].tolist()
urls = [i.split('?')[0].split('/')[-1] for i in urls]
urls = list(set(urls))
print(urls)
walls_id = []
import re

def parse_wall_id(wall_id):
    match = re.match(r'wall(-?\d+)_(\d+)', wall_id)
    if match:
        owner_id = match.group(1)
        post_id = match.group(2)
        return owner_id, post_id
    else:
        raise ValueError("Invalid wall id format")
for wall_id in urls:
    try:
        owner_id, post_id = parse_wall_id(wall_id)
        walls_id.append((owner_id, post_id))
        print(f"Owner ID: {owner_id}, Post ID: {post_id}")
    except ValueError as e:
        print(e)
```

```python
access_token = 'b4b62039b4b62039b4b6203966b7af23b0bb4b6b4b62039d2cb3123fe57d65f35493dce'
```

```python
import requests
import time

def get_comments(post_id, owner_id, access_token, version='5.131'):
    base_url = 'https://api.vk.com/method/'
    method = 'wall.getComments'
    params = {
        'owner_id': owner_id,
        'post_id': post_id,
        'access_token': access_token,
        'v': version,
        'count': 100,  
    }

    all_comments = []
    while True:
        time.sleep(0.34)
        response = requests.get(base_url + method, params=params).json()
        if 'error' in response:
            print(f"Error: {response['error']['error_msg']}")
            break

        comments = response['response']['items']
        all_comments.extend(comments)

        if len(comments) < params['count']:
            break  # Выход из цикла, если комментарии закончились
        # Обновляем смещение для следующего запроса
        params['offset'] = params.get('offset', 0) + params['count']

    return all_comments
```

```python
from tqdm import tqdm

df_comments = []
for owner_id, post_id in tqdm(walls_id, desc="Хаваем url"):
    comments = get_comments(post_id, owner_id, access_token)
    for comment in comments:
            df_comments.append({
                'owner_id': owner_id,
                'post_id': post_id,
                'comment_id': comment['id'],
                'from_id': comment['from_id'],
                'text': comment['text'],
                'date': comment['date'],
            })
df = pd.DataFrame(df_comments)
```
