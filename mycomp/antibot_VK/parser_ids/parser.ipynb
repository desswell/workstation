```python
import vk_api
import pandas as pd
```

```python

def get_friends_ids(api, user_id):
    try:
        friends = api.friends.get(user_id=user_id)
        return friends['items']
    except vk_api.exceptions.ApiError as e:
        print(f"Error fetching friends for user {user_id}: {e}")
        return []
```

```python

def get_all_ids(api, initial_user_ids):
    all_ids = set()
    to_process = initial_user_ids

    for i in range(3):  # три колена
        next_to_process = []
        for user_id in to_process:
            if user_id not in all_ids:
                all_ids.add(user_id)
                friends_ids = get_friends_ids(api, user_id)
                next_to_process.extend(friends_ids)
        to_process = next_to_process

    return list(all_ids)

```

```python
def main():
    # Ваш токен API ВКонтакте
    token = 'bb01e286bb01e286bb01e2860ab81977f2bbb01bb01e286dd647517a5bcfa4df0bccfcd'
    
    # Инициализация сессии
    vk_session = vk_api.VkApi(token=token)
    vk = vk_session.get_api()
    
    # Ваши начальные ID пользователей
    initial_user_ids = ['254426082', '615482595', '335178058']
    
    all_ids = get_all_ids(vk, initial_user_ids)
    
    # Создание DataFrame
    df = pd.DataFrame(all_ids, columns=['user_id'])
    
    # Сохранение в CSV
    df.to_csv('vk_user_ids.csv', index=False)
    
    print(f"Total number of unique IDs collected: {len(all_ids)}")
    print(df)

if __name__ == '__main__':
    main()
```
